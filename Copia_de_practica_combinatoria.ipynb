* son sin repetición ya que si uno ha resultadi ser presidente, no puede ser simultaneamente tesorero o secretario.

##Práctica 1.3
Invente un caso de **variaciones sin repetición**.

¿Cuántos números de 3 cifras se puden escribir con los dígitos 1,2,3,4,5 sin poderse repetir?




### Solución 1.3
Variaciones sin repetición de 5 elementos tomados de 3 en 3.

$$V^r_n=\frac{n!}{(n-r)!}=\frac{5!}{(5-3)!}=\frac{5!}{2!}=\frac{5 \cdot 4 \cdot 3 \cdot 2!}{2!}=5 \cdot 4 \cdot 3=60$$

Sabemos que son Variaciones sin repetición ya que:
* No tomo todos los elementos ya que cojo 3 de 5.
* el orden importa, ya que no es lo mismo el numero 125 que el 512 o el 251.
* son sin repetición ya que si un número ya ha salido no se puede volver a repetir.


```python
import itertools
it = itertools.permutations(['1','2','3','4','5'],3)
li = list(it)
print(li)

print()

for i in range(len(li)):
    #print("Variación", i+1,":", *li[i])
    #print(f"Variación {i+1:2d}:  {li[i]}")
    print(f"Variación {i+1:2d}: ", *li[i])
```

## Práctica 1.4
Invente un caso de **variaciones con repetición**.     
¿Cuántas combinaciones distintas existen para un candado con tres ruletas numeradas del 1 al 9 números cada una? 

### Solución 1.4
Las variaciones con repetición de m elementos tomados de n en n tiene como fórmula:


\(VR 9^3 = 729\)  
Sabemos que son Variaciones con repetición ya que:
* No tomo los 9 números porque el canadado solo tiene 3 ruletas
* el orden importa, ya que no es la misma combinación 351 que 513 o 153.
* son con repetición ya que todas las ruletas tienen los números del 1 al 9.

```python
import itertools

 

def variaciones_con_repeticion(n,seq):

    contador = 0

    for p in itertools.product(seq,repeat=n):

      contador += 1

      print(contador, "".join(p))

 

secuencia = '123456789'

m = len(secuencia)

n = 3

variaciones_con_repeticion(n, secuencia)

 

print(f"El numero de variaciones con repetición de {m} elementos tomados de {n} en {n} son {m**n}. ")
```

## Práctica 1.5
Invente un caso de **permutaciones sin repetición**.    
Juan (A), pedro (B), Luis (C) y Carlos (D) van al cine esta tarde ¿De cuantás maneras diferentes pueden sentarse en cuatro asientos juntos?

### Solución 1.5
* Dados $n$ elementos, se pueden obtener $n!$ posibles formas de reordenarlos, esto es, existen $n!$ permutaciones.
* Si $n=4$ existen  
$$4! = 4 \cdot 3 \cdot 2 \cdot 1 = 24$$
Sabemos que son permutaciones sin repetición ya que:
* Primero llega Juan y ve que tiene las cuatro plazas libres, luego tiene **4** posiblidades para poder sentarse. Se sienta en una de ellas.
* En segundo lugar llega Pedro  y ve que solo tiene **3** plazas libres para sentarse y se sienta en una de ellas.
* Después, llega Luis  y ve que únicamente tiene **2** plaza libre para aparcar, y se sienta en una de ellas.
* Finalmente llega Carlos y ve que únicamente tiene **1** plaza libre para sentarse, y se sienta en ella.
* Aplicando el principio fundamental del conteo vemos que las permutaciones de ABCD se calculan multiplicando las posibilidades de cada uno de ellos que son:
 - 4 * 3 * 2 * 1 = 24
 - que podemos expresar como 4!

```python
from random import seed, sample
from math import factorial
seed()
matriz = []

lista = ["a", "b", "c", "d"]
n = len(lista)

for _ in range(factorial(n)):
    while True:
        candidata = sample(lista, n)
        if candidata not in matriz:
            matriz.append(candidata)
            #print(matriz)
            break

matriz.sort()
for i in range(len(matriz)):
    print(*matriz[i])


```

### Solución 1.9
Son 5 personas que se han de sentar en una mesa redonda: PABCD, siendo P el presidente y D el ministro de defensa.
 La fórmula sería:$$5-1! = 4 \cdot 3 \cdot 2 \cdot 1 = 24$$
P y D siempre van juntos. O bien, se sientan como PD o como DP, esto es, el ministro de defensa se sienta o a un lado o al otro del presidente.

Podemos fijar PD como posición fija, por ejemplo, en la primera posición y luego permutar el reto de asistentes que son ABC. De esta forma obtendríamos:

PD ABC

PD ACB

PD BAC

PD BCA

PD CAB

PD CBA

Ahora consideremos el otro caso, cuando el ministro de defensa se sienta al otro lado del presidente. Los casos obtenidos serían:

DP ABC

DP ACB

DP BCA

```python
import math
from itertools import permutations

p = list(permutations('ABC'))   # lista de permutaciones de 3 elementos



texto = ''   # inicializamos una variable texto que es un string vacío que luego iremos concatenando

for v in p:

    texto += 'PD' + ''.join(v) + ','   # El ministro de Defensa (D) se sienta a un lado del Presidente (P)

for v in p:

    texto += 'DP' + ''.join(v) + ','   # El ministro de Defensa (D) se sienta al otro lado del Presidente (P)



print(texto[:-1])   
```

La fórmula que permite hallar el valor de un número combinatorio es la siguiente: ( m n) = m! n! ⋅ ( m − n)!
